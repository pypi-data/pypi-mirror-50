```python
import sys
sys.path.append('..')

from sklearn.datasets import load_diabetes
import numpy as np
from korr import slice_yx, pearson, corrgram
```

```python
y = load_diabetes().target
X = load_diabetes().data
```

```python
r, pval = pearson(np.c_[y, X])
```

```python
y_r, y_pval, x_r, x_pval = slice_yx(r, pval, ydim=1)
```

```python
print(np.c_[y_r, y_pval].round(3))
```

```python
corrgram(x_r, x_pval);
```
